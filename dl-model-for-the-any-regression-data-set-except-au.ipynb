```python
import keras
import pandas as pd 
```

```python
dataset_path=keras.utils.get_file("auto_mpg.data","https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
```

```python
column_names=["MPG","Cylinders","Dispalcement","Horsepower","Weight",
             "Accelerartion","Model Year","Origin"]
data=pd.read_csv(
    dataset_path,names=column_names,
    comment="\t",
    na_values="?",
    sep=" ",
    skipinitialspace=True
                )
data.head(20)
```

```python
data_copy=data.copy()
data_copy.head()
```

```python
data_copy.tail()
```

```python
data_copy.isna()
```

```python
data_copy.isna().sum()
```

```python
data_copy=data_copy.dropna()
```

```python
data_copy.isna().sum()
```

```python
origin=data_copy.pop("Origin")
```

```python
data_copy.tail()
```

```python
data_copy["USA"]=(origin==1)*1.0
data_copy["Europe"]=(origin==2)*1.0
data_copy["Japan"]=(origin==3)*1.0
```

```python
data_copy.tail()
```

```python
train_data_copy=data_copy.sample(frac=0.8)
test_data_copy=data_copy.drop(train_data_copy.index)
```

```python
train_data_copy.tail()
```

```python
test_data_copy.tail()
```

```python
train_labels=train_data_copy.pop("MPG")
test_labels=test_data_copy.pop("MPG")
```

```python
train_data_copy.head()
```

```python
test_data_copy.head()
```

```python
train_status=train_data_copy.describe()
train_status
```

```python
train_status=train_data_copy.describe().T
train_status
```

```python
def nor(X):
    return (X-train_status["mean"])/(X-train_status["std"])
```

```python
nor_train=nor(train_data_copy)
nor_train
```

```python
nor_test=nor(test_data_copy)
nor_test
```

```python
from tensorflow.keras import layers
import tensorflow as tf

def build_model():
    model=keras.Sequential([
        layers.Dense(64,activation="relu",input_shape=[len(train_data_copy.keys())]),
        layers.Dense(64,activation="relu"),
        layers.Dense(1)
    ])
    optimizer=tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss="mse",
                 optimizer=optimizer,
                  metrics=["mae","mse"]
                 )
    return model
model=build_model()
```

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self,epoch,logs):
        if epoch % 10==0:print("")
        print(".",end="")

EPOCH=1000
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
    hist=pd.DataFrame(history.history)
    hist["epoch"]=history.epoch
    
    plt.figure()
    plt.xlabel("Epoch")
    plt.ylabel("Mean abs Error [MPG]")
    plt.plot(hist["epoch"],hist["mae"],
            label="Train Error")
    plt.plot(hist["epoch"],hist["val_mae"],
            label="val Error")
    plt.ylim([0,5])
    plt.legend()
    #######
    plt.figure()
    plt.xlabel("Epoch")
    plt.ylabel("Mean Square Error [$MPG]")
    plt.plot(hist["epoch"],hist["mse"],
            label="Train Error")
    plt.plot(hist["epoch"],hist["val_mae"],
            label="val Error")
    plt.ylim([0,20])
    plt.legend()
    
    plt.show()
```

```python
early_stop=keras.callbacks.EarlyStopping(monitor="val_loss",patience=10)

history=model.fit(nor_train,train_labels,epochs=EPOCH,
                 validation_split=0.2,verbose=0,callbacks=[early_stop,PrintDot()])

plot_history(history)
loss,mae,mse=model.evaluate(nor_test,test_labels,verbose=0)
```
